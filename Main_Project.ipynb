```python
#importing necesary packaged
import json
import pandas as pd

from ari import ari_to_features
import pandas as pd
from tqdm import tqdm
import re
```

```python
#Loading the first json file, with the forst 1000 playlist into a variable
with open('Data/spotify_million_playlist_dataset/Main_data/mpd.slice.0-999.json') as f:
    data = json.load(f)
```

```python
#running a loop to save it all into one dataframe
# Create an empty list to store the flattened playlists
flattened_playlists = []

# Iterate over each playlist
for playlist in data['playlists']:

    # Iterate over each track in the playlist
    for track in playlist['tracks']:

        # Create a new dictionary to store the flattened data
        flattened_data = {}

        # Add the playlist-level data to the flattened dictionary
        flattened_data['pid'] = playlist['pid']
        flattened_data['name'] = playlist['name']
        flattened_data['collaborative'] = playlist['collaborative']
        flattened_data['modified_at'] = playlist['modified_at']
        flattened_data['num_tracks'] = playlist['num_tracks']
        flattened_data['num_albums'] = playlist['num_albums']
        flattened_data['num_followers'] = playlist['num_followers']

        # Add the track-level data to the flattened dictionary
        flattened_data['pos'] = track['pos']
        flattened_data['artist_name'] = track['artist_name']
        flattened_data['track_uri'] = track['track_uri']
        flattened_data['artist_uri'] = track['artist_uri']
        flattened_data['track_name'] = track['track_name']
        flattened_data['album_uri'] = track['album_uri']
        flattened_data['duration_ms'] = track['duration_ms']
        flattened_data['album_name'] = track['album_name']

        # Add the flattened data to the list
        flattened_playlists.append(flattened_data)

# Create a DataFrame from the flattened list of playlists
df = pd.DataFrame(flattened_playlists)
```

```python
def drop_duplicates(df):
    '''
    Drop duplicate songs
    '''
    df['artists_song'] = df.apply(lambda row: row['artist_name']+row['track_name'],axis = 1)
    return df.drop_duplicates('artists_song')

songDF = drop_duplicates(df)
print("Are all songs unique: ",len(pd.unique(songDF.artists_song))==len(songDF))
```

```python
#label encoding all categorical variables first
#storing all the "objects" in a new list
colname = []
for x in df.columns:
    if df[x].dtype == 'object':
        colname.append(x)
colname
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

for x in colname:
    df[x]= le.fit_transform(df[x])
    le_name_mapping = dict(zip(le.classes_,le.transform(le.classes_)))
    print("feature", x)
    print("mapping", le_name_mapping)
```

```python
#now making the matrix
iimatrix = pd.DataFrame(index=df.track_uri.unique(), columns=df.columns[6:])
```

```python
for track_uri in iimatrix.index:
    iimatrix.loc[track_uri] = df[df['track_uri'] == track_uri].iloc[:, 6:].mean().values
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
# Calculate cosine similarity between all pairs of songs
cos_sim_matrix = cosine_similarity(iimatrix)
```

```python
print(cos_sim_matrix)
```

```python
# Convert the cosine similarity matrix into a Pandas DataFrame
cos_sim_df = pd.DataFrame(cos_sim_matrix, index=iimatrix.index, columns=iimatrix.index)
```

```python
cos_sim_df.head()
```

```python
# Define a function to get recommendations for a playlist based on cosine similarity
def get_reccs(pid, n=10):
    # Get the cosine similarity scores for the input playlist
    sim_scores = cos_sim_df.loc[pid]
    
    # Get the indices of the playlists with the highest similarity scores
    top_playlists = sim_scores.nlargest(n+1).index[1:]
    
    # Get the track URIs for the top playlists
    track_uris = df[df['pid'].isin(top_playlists)]['track_uri']
    
    # Get the track details for the recommended tracks
    rec_songs = df[df['track_uri'].isin(track_uris)].drop_duplicates('track_uri')[['track_uri', 'track_name', 'artist_name']]
    
    
    return rec_songs
    

```

```python
get_reccs(1,10)
```

```python
# Get recommendations for playlist ID 1
recs = get_reccs(1, 10)

# Inverse transform the track URI column
recs_uris = le.inverse_transform(recs['track_uri'].values)
recs_artists = le.inverse_transform(recs['artist_name'].values)
recs_album = le.inverse_transform(recs['album_name'].values)

# Combine the inverse transformed track URIs with the other columns
recs1 = pd.DataFrame({'track_name': recs_uris, 'artist_name': recs_artists, 'album_name': recs_album})

# Print the recommendations
print(recs1)
```
